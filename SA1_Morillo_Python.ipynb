```python
import random

def ran_word():
    with open("words.txt", "r") as f:
        words = f.readlines()
    return random.choice(words).strip()

def dis_word(word, guesses):
    display = ""
    for letter in word:
        if letter in guesses:
            display += letter + " "
        else:
            display += "_ "
    return display

def user_guess():
    print("\n")
    guess = input("Enter your guess: ")
    if len(guess) != 1:
        print("You can only enter a single letter.")
        return user_guess()
    elif not guess.isalpha():
        print("You can only enter a letter.")
        return user_guess()
    else:
        return guess

def main_game():
    word = ran_word()
    guesses = set()
    incorrect_guesses = 0
    max_incorrect_guesses = 8

    while True:
        print(dis_word(word, guesses))
        if "_" not in dis_word(word, guesses):
            print("\n")
            print("Congratulations, you got it!")
            break
        elif incorrect_guesses == max_incorrect_guesses:
            print("\n")
            print(f"Oops! You ran out of guesses. The word was {word}. Better luck next time!")
            break
        else:
            guess = user_guess()
            if guess in guesses:
                print("\n")
                print("Letter have been guessed. Try another letter.")
            elif guess in word:
                guesses.add(guess)
                print("\n")
                print("Great guess!")
            else:
                guesses.add(guess)
                incorrect_guesses += 1
                print("\n")
                print(f"Unfortunately, {guess} is not part of the word. You have {max_incorrect_guesses - incorrect_guesses} guesses left.")

if __name__ == '__main__':
    main_game()
```

```python
# Number 2
import time
#iteration
print("iteration")
ite_time = time.time()
def fib_iter(n):
    a=1
    b=1
    if n==1:
        print('0')
    elif n==2:
        print('0','1')
    else:
        print('0',a,b,end=' ')
        for i in range(n-3):
            total = a + b
            b=a
            a= total
            print(total,end=' ')
        print()
        return b
         
fib_iter(15)
print("The time rendered in recursion is :",(time.time()-ite_time), "seconds")
    
print("\n")

#recursion
print("recursion")
rec_time = time.time()
def recur_fibo(n):  
   if n <= 1:  
       return n  
   else:  
       return(recur_fibo(n-1) + recur_fibo(n-2)) 
nterms = 15
for i in range(nterms):  
    print(recur_fibo(i)) 

print("The time rendered in recursion is :",(time.time()-rec_time), "seconds")

if ite_time < rec_time:
    print("Iteration is faster than recursion")
elif ite_time > rec_time:
    print("Iteration is slower than recursion")
else:
    print("Iteration is as fast as recursion")
```
